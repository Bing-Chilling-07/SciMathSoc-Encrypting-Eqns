Functions that will come in use in algorithm

```python
# Implement padding for encrypting my text
def pad(text):
    pad_len = 16 - (len(text) % 16)
    return text + bytes([pad_len] * pad_len)



# To unpad the ciphertext while decryption process
def unpad(padded_text):
    pad_len = padded_text[-1]
    return padded_text[:-pad_len]



# XOR operation in bytes
def xor_bytes(a, b):
   result = []
   for i in range(len(a)):
    result.append(a[i] ^ b[i])  # XOR bytes at the same position
   return bytes(result)
```

Defining Round Constants(needed  for Key Expansion)

```python
Round_constants = [0x01, 0x02, 0x04, 0x08, 0x10, 0x20, 0x40, 0x80, 0x1B, 0x36]
# Values in hexadecimal
```

Matrix to Bytes conversion and Vice-Versa

```python
def bytes_to_matrix(text):
    # Convert bytes to matrix(4x4 in our case)
    matrix = []
    for i in range(0, len(text), 4):
        row = list(text[i:i+4])
        matrix.append(row)
    return matrix


def matrix_to_bytes(matrix):
    # Flatten the 4x4 matrix into a single list of bytes and convert to bytes
    byte_list = []
    for row in matrix:
        byte_list.extend(row)
    return bytes(byte_list)
```

Defining Operations for the AES Encryption,\
For this we will create a class.

```python
class AES:


    def __init__(self, key):
        assert len(key) == 16
        # Ensure that key is 16 bytes

        self.round_keys = self.key_expansion(key)




    # Key expansion: extending number of  rounds to 11

    def key_expansion(self, key):

        # Convert the intial key to  matrix (4 x 4)
        key_columns = bytes_to_matrix(key)

        # Copy the matrix into round key array
        round_keys = key_columns.copy()


# For the susequent words:

        for i in range(4, 44):

            word = list(round_keys[i - 1])

            # Segregating into entities of 4 words
            if i % 4 == 0:

              # First rotating the word, then substituting it using the S-Box
                word = self.sub_word(self.rotation_word(word))

                # Finally adding the round constants [XOR]
                word[0] ^= Round_constants[i // 4 - 1]

            xor_word = []

            for j in range(4):

               # XOR  addition of jth column of previous word
                xor_word.append(word[j] ^ round_keys[i - 4][j])
            round_keys.append(xor_word)



        grouped_keys = []



        for i in range(0, len(round_keys), 4):

          # Keys in 4 consecutive rounds
            grouped_keys.append([round_keys[i], round_keys[i+1], round_keys[i+2], round_keys[i+3]])

        return grouped_keys





# Word Rotation: Done for Key Expansion

    def rotation_word(self, word):

      # shifting by 1 byte to get a permutation
        return word[1:] + word[:1]









# Substituted words : Done for Key Expansion

    def sub_word(self, word):
        substituted = []
        for byte in word:

          # Using AES-S Box to get a new state array
            substituted.append(SBOX[byte])

        return substituted








# Adding( through XOR operation) Round Key

    def add_round_key(self, state, key):
        new_state = []
        for i in range(4):
            row = []
            for j in range(4):

              # Adding to rows
                row.append(state[i][j] ^ key[i][j])

             # Adding new rows to the state array
            new_state.append(row)

        return new_state







# Obtaining sub-bytes, required for process of encryption

    def sub_bytes(self, state):

        new_state = []


        for row in state:

            new_row = []

            for byte in row:

              # adding S-BOX transformation to rows of  state array
                new_row.append(SBOX[byte])

            # Add the whole row to the state array
            new_state.append(new_row)

        return new_state






# Inverting the sub-bytes to get the plaintext for decryption process

    def inv_sub_bytes(self, state):

        new_state = []

        for row in state:

            new_row = []

            for byte in row:

              # Using the Inverse S-BOx transformation to retrieve the original bytes
                new_row.append(INV_SBOX[byte])

            # Add the whole row to the inverted state array
            new_state.append(new_row)

        return new_state






# Shifting rows, part of permutations (For encryption)

    def shift_rows(self, state):

        new_state = []

        for i in range(4):

          # Circular Shifting in i th row to left by i blocks
            row = state[i][i:] + state[i][:i]

            new_state.append(row)
        return new_state





# Invert shifting (For decryption)

    def inv_shift_rows(self, state):

        new_state = []

        for i in range(4):

          # Circular Shifting in i th row to right by i blocks
            row = state[i][-i:] + state[i][:-i]

            new_state.append(row)
        return new_state








# Mixing of columns (For encryption)

# [d0 d1 d2 d3].T  =    [2 3 1 1
#                        1 2 3 1     [b0 b1 b2 b3].T
#                        1 1 2 3
#                        3 1 1 2]
# Each column of the new state array is caluclated through the transformation
# matrix using the corresponding columns of the old state array



    def mix_columns(self, state):

      # Initializing the list
        mixed = [[0] * 4 for i in range(4)]

        for c in range(4):
            a = [state[row][c] for row in range(4)]
            mixed[0][c] = self.Galois_multiplication(a[0], 2) ^ self.Galois_multiplication(a[1], 3) ^ a[2] ^ a[3]
            # Zeroth row entry: d0= 2 x b0 + 3 x b1 + 1 x b3 + 1 x b4 using Galois multiplication

            mixed[1][c] = a[0] ^ self.Galois_multiplication(a[1], 2) ^ self.Galois_multiplication(a[2], 3) ^ a[3]
            # First row entry: d1= 1 x b0 + 2 x b1 + 3 x b2 + 1 x b3

            mixed[2][c] = a[0] ^ a[1] ^ self.Galois_multiplication(a[2], 2) ^ self.Galois_multiplication(a[3], 3)
            # Second row entry: d2= 1 x d0 + 1 x d1 + 2 x d2 + 3 x d3

            mixed[3][c] = self.Galois_multiplication(a[0], 3) ^ a[1] ^ a[2] ^ self.Galois_multiplication(a[3], 2)
            # Third row entry: d3= 3 x b0 + 1 x b1 +1 x b2 + 2 x b3

        return mixed







# Invert mixing (For decryption)

    def inv_mix_columns(self, state):

      # Initializing the list
        mixed = [[0] * 4 for i in range(4)]

      # Invert the mixing operations
        for c in range(4):
            a = [state[row][c] for row in range(4)]

            # Carrying out all inverse operations:

            mixed[0][c] = self.Galois_multiplication(a[0], 0x0e) ^ self.Galois_multiplication(a[1], 0x0b) ^ self.Galois_multiplication(a[2], 0x0d) ^ self.Galois_multiplication(a[3], 0x09)

            mixed[1][c] = self.Galois_multiplication(a[0], 0x09) ^ self.Galois_multiplication(a[1], 0x0e) ^ self.Galois_multiplication(a[2], 0x0b) ^ self.Galois_multiplication(a[3], 0x0d)

            mixed[2][c] = self.Galois_multiplication(a[0], 0x0d) ^ self.Galois_multiplication(a[1], 0x09) ^ self.Galois_multiplication(a[2], 0x0e) ^ self.Galois_multiplication(a[3], 0x0b)

            mixed[3][c] = self.Galois_multiplication(a[0], 0x0b) ^ self.Galois_multiplication(a[1], 0x0d) ^ self.Galois_multiplication(a[2], 0x09) ^ self.Galois_multiplication(a[3], 0x0e)


        return mixed



# Defining Galois multiplication

    def Galois_multiplication(self, a, b):

      # Basically multiplying under Galois field of 2^8(2^8 for our algorithm)

        result = 0

        for iterations in range(8):

            if b & 1:
                result ^= a
            high_bit = a & 0x80

            # shifting to the next bit
            a <<= 1

            if high_bit:
                a ^= 0x1b
            b >>= 1
        return result % (2**8)



# Encyrption:

    def Encryption(self, plaintext):

        assert len(plaintext) == 16
        # Ensuring key to be of 16 bytes


# ROUND 0:

        # Converting to bytes
        state = bytes_to_matrix(plaintext)


        # Adding (XOR operation) round key
        state = self.add_round_key(state, self.round_keys[0])


        # In zeroth round , no need of mixing columns




# ROUND 1-9:

        for i in range(1, 10):

          # Transforming to sub bytes
            state = self.sub_bytes(state)

          # Shifting rows
            state = self.shift_rows(state)

          # Mixing of columns
            state = self.mix_columns(state)

          # Adding the Round Key (XOR)
            state = self.add_round_key(state, self.round_keys[i])



# ROUND-10:

       # Transforming to sub-bytes
        state = self.sub_bytes(state)

        # Row-shifting
        state = self.shift_rows(state)

        #Adding round key(XOR)
        state = self.add_round_key(state, self.round_keys[10])

        # No mixing of columns is required in the last(10th) round as well



        # Final encrypted text:
        result = matrix_to_bytes(state)
        return result







# Decryption:

     # The structure for Decryption function is quite to that of Encryption function,its just that we have to invert all the operations

    def Decryption(self, ciphertext):

        assert len(ciphertext) == 16
        # Again 16-byte length is to be ensured



# ROUND 10 inversion :

        # Converting to bytes
        state = bytes_to_matrix(ciphertext)

        # Adding round keys(inversion)
        state = self.add_round_key(state, self.round_keys[10])

        # Invert row-shifting operations
        state = self.inv_shift_rows(state)

        # Inverting sub-byte transformation
        state = self.inv_sub_bytes(state)

        # No need for inversion of mixing as no mixing was done in this round





# ROUND 1-9 inversion ( in reverse order,i.e, 9->8->7...->1):

        for i in range(9, 0, -1):

            # Adding round keys(inversion)
            state = self.add_round_key(state, self.round_keys[i])

            # Invert mixing of columns
            state = self.inv_mix_columns(state)

            # Invert row-shifting operations
            state = self.inv_shift_rows(state)

            # Inverting sub-byte transformation
            state = self.inv_sub_bytes(state)



#ROUND 0:
        state = self.add_round_key(state, self.round_keys[0])


# Decrypted text:
        result = matrix_to_bytes(state)
        return result
```
